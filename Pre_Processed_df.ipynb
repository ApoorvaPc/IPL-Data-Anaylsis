```python
# import pandas 
# import two Datasets - Matches and Ball-by-Ball Dataset

import pandas as pd 

Matches_raw_data = pd.read_csv("IPL_Matches_2008-2020.csv")
Ball_by_Ball_raw_data = pd.read_csv("IPL_Ball-by-Ball_2008-2020.csv")
```

```python
# Making copies of Original DataFrame

Matches_df = Matches_raw_data.copy()       
Ball_by_Ball_df = Ball_by_Ball_raw_data.copy()  
```

```python
# Preprocessing of Matches_df

Matches_df
```

```python
# Remove Unwanted Columns

Matches_df.drop(['neutral_venue','result_margin', 'method', 'umpire1', 'umpire2'], axis = 1, inplace=True)
```

```python
# Check columns and  Null values in all Columns

Matches_df.info()
```

```python
# Display Rows having Null values in City column

Matches_df[Matches_df.city.isnull()]
```

```python
# Fill missing values in City based on Venue

Matches_df.loc[Matches_df['venue'] == 'Dubai International Cricket Stadium', 'city' ] = 'Dubai'
Matches_df.loc[Matches_df['venue'] == 'Sharjah Cricket Stadium', 'city' ] = 'Sharjah'      
```

```python
# Check Null values in City Column

Matches_df[Matches_df.city.isnull()]
```

```python
# No null values found in City 
```

```python
#  Fill Matches which resulted in a Tie using fillna()

Matches_df.fillna({
    'winner' : 'tie',
    'result' : 'tie',
    'eliminator' : 'Not allowed',
    'player_of_match' : 'Not allowed'
}, inplace=True)     
```

```python
# Check Null Values in all Columns

Matches_df.info()
```

```python
# No Null values in any Column
```

```python
# Convert 'date' into datetime type

Matches_df['date'] = pd.to_datetime(Matches_df['date'], format = '%m/%d/%Y')
```

```python
Matches_df['date']
```

```python
# Extract Only the Year value from datetime

Matches_df['date'][0].year
```

```python
# Create list to hold all the Years

list_years = []
list_years
```

```python
# Filling the list using for loop

for i in range(Matches_df.shape[0]):
    list_years.append(Matches_df['date'][i].year)
        
```

```python
# Check the Year Column

Matches_df.head(10)
```

```python
# Drop date column

Matches_df.drop(['date'], axis = 1, inplace = True)
```

```python
# Rename the Colums

column_names = ['ID', 'City', 'Player_of_the_Match', 'Venue', 'Team_1', 'Team_2',
       'Toss_winner', 'Toss_decision', 'Winner', 'Result', 'Eliminator', 'Year']
Matches_df.columns = column_names
```

```python
# Building a Bridge between Python and Mysql using pymysql

import pymysql               
```

```python
# Using a variable to Designate the Database we would like to connect to

conn = pymysql.connect(database = 'IPL_database', user = 'root', password = 'password')
```

```python
# Create a variable called cursor and invoke cursor method to indicate where to write the code in Mysql

cursor = conn.cursor()
```

```python
# Create a Structure that will insert information of both 'IPL_Matches' and 'IPL_Ball_by_Ball' Dataframe into MySQL Database 'IPL_database'
# This is done by using 'one multi-row insert statement'.
```

```python
# To run SQL code in Python we need to apply execute method to the cursor object

cursor.execute('SELECT * FROM IPL_Matches;')
```

```python
# Creating insert_query variable to insert values

insert_query = 'INSERT INTO IPL_Matches VALUES'
```

```python
insert_query
```

```python
# Check the shape of the dataframe to use in our 'for loop'

IPL_Matches.shape
```

```python
# Using for loop which will iterate over both rows and columns using the shape method to extract each value from the Dataframe
# Seperating values with ',' and adding '" "' to column values containing string

for i in range(IPL_Matches.shape[0]):
    insert_query += '('
    
    for j in range(IPL_Matches.shape[1]):
        if j == 0 or j == 2:
            insert_query += str(IPL_Matches[IPL_Matches.columns.values[j]][i]) + ', '
        else:
            insert_query += '"' + (IPL_Matches[IPL_Matches.columns.values[j]][i]) + '", '
        
    insert_query = insert_query[:-2] + '), '   
        
```

```python
# End the query with ';' to complete the SQL insert query by overwriting the code using Reverse Indexing 

insert_query = insert_query[:-2] + ';'
```

```python
insert_query
```

```python
# Execute the query in MySQL by using the variable insert_query as an argument of execute method

cursor.execute(insert_query)
```

```python
# Use Commit Method to make Changes to the Database in MySQL

conn.commit()
```

```python
# Check Contents of IPL_Ball_by_Bll Datatable in MySQL

cursor.execute('SELECT * FROM IPL_Ball_by_Ball;')
```

```python
# Creating the Insert Statment for the Second table

second_insert_query = 'INSERT INTO IPL_Ball_by_Ball VALUES'         
```

```python
second_insert_query
```

```python
# The number of rows to be inserted in MySQL using single query is limited to 20000 rows.
# Since we have 193468 rows, we'll divide it into 11 seperate multi-row insert_queries
# Using for loop we will iterate 10 times to insert all values


a = 0
b = int(IPL_Ball_by_Ball.shape[0]/11)

# Using list to hold columns containing string values so we can add '" "'
list = [0, 1, 2, 3, 7, 8, 9, 10]
```

```python
# Using for loop which will iterate over both rows and columns using the shape method to extract each value from the Dataframe


for i in range(11):
    second_insert_query = 'INSERT INTO IPL_Ball_by_Ball VALUES'         
    for k in range(a, a + b ,1):
        second_insert_query += '('
       
        for l in range(IPL_Ball_by_Ball.shape[1]):
            if l in list:
                second_insert_query += str(IPL_Ball_by_Ball[IPL_Ball_by_Ball.columns.values[l]][k]) + ', '
            else:
                second_insert_query += '"' + str(IPL_Ball_by_Ball[IPL_Ball_by_Ball.columns.values[l]][k]) + '",'
        second_insert_query = second_insert_query[:-1] + '), '  
    second_insert_query = second_insert_query[:-2] + ';' 
    cursor.execute(second_insert_query)
    a = a + b 
```

```python
# Use Commit Method to make Changes to the Database in MySQL

conn.commit()
```

```python
# Check Contents of IPL_Ball_by_Bll Datatable in MySQL

cursor.execute('SELECT * FROM IPL_Ball_by_Ball;')
```
